```python
def stats(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = newData.apply(stats)
loc_stats
```

Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. 
Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
import numpy as np
MD = pd.read_csv('us-marriages-divorces-1867-2014.csv')
y=MD.Marriages_per_1000
y1=MD.Divorces_per_1000
x=MD.Year
plt.plot(x,y)
plt.plot(x,y1)
columns = MD.columns.drop(['Year'])
plt.legend(columns)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of marriages and divorces per capita in the U.S')
plt.show()
```

Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
MD1 = pd.read_csv('us-marriages-divorces-1867-2014.csv')
MD2=MD1[MD1['Year']==1900]
MD3=MD1[MD1['Year']==1950]
MD4=MD1[MD1['Year']==2000]
frames = [MD2, MD3, MD4]
Sub_MD = pd.concat(frames)
Sub_MD
Sub_MD.plot(x="Year", y=['Marriages', 'Divorces'], kind="bar")
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Marriages VS Divorces in 1900,1950 and 2000')
plt.show()
```

Question 9
Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt
RE= pd.read_csv('roman-emperor-reigns.csv')
A=RE.Cause_of_Death.where(RE.Cause_of_Death=='Assassinated').count()
O=RE.Cause_of_Death.count()
A_Fraction=(A/O)*100
O_Fraction=100-A_Fraction
Data=[A_Fraction,O_Fraction]
F_labels=('Assassinated','Other')
plt.pie(Data,labels=F_labels, autopct='%1.1f%%')
plt.title('Fraction of assassinated to With Total Deaths')
plt.show()
```

Question 10
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
ARCD = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
sns.scatterplot(x='Year', y='Computer Science Doctorates Awarded (US)',hue='Year',data=ARCD)
sns.scatterplot(x='Year', y='Total Arcade Revenue (billions)', hue='Year', data=ARCD)
plt.title('Relationship between total revenue earned by arcades and the number of Computer Science PhDs')
plt.xlabel('Year')
plt.ylabel('CSDA(US)&TAR(billions)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```
